# analise do eleitorado brasileiro


```python
import pandas as pd

df = pd.read_csv('dados/Emunicipio2018.csv', sep=';', encoding='latin1')

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.head().T
```

```python
df.nlargest(5,'QTD_ELEITORES_DEFICIENTE')
```

```python
print('Eleitoras: ', df['QTD_ELEITORES_FEMININO'].sum())
print('Eleitores: ', df['QTD_ELEITORES_MASCULINO'].sum())
print('N informado: ', df['QTD_ELEITORES_NAOINFORMADO'].sum())
```

```python
total = df['QTD_ELEITORES'].sum()
feminino = df['QTD_ELEITORES_FEMININO'].sum()
masculino = df['QTD_ELEITORES_MASCULINO'].sum()
nanofor = df['QTD_ELEITORES_NAOINFORMADO'].sum()
```

```python
print('Eleitoras: ', ((feminino / total)*100).round(2))
print('Eleitores: ', ((masculino / total)*100).round(2))
print('N informado: ', ((nanofor / total)*100).round(2))
```

```python
df[df['QTD_ELEITORES_MASCULINO']>df['QTD_ELEITORES_FEMININO']].count()
```

```python
#criacao da coluna relacao feminino / masculino
df['RELACAOFEM'] = df['QTD_ELEITORES_FEMININO'] / df['QTD_ELEITORES_MASCULINO'] 
```

```python
df.nlargest(5,'RELACAOFEM').T
```

```python
df.nsmallest(5,'RELACAOFEM').T
```

```python
df_brasil = df[df['CD_PAIS'] == 1].copy()

df_brasil.head(3)
```

```python
df_brasil.nlargest(5,'RELACAOFEM').T
```

```python
df_brasil.nsmallest(5,'RELACAOFEM').T
```

```python
%matplotlib inline
df_brasil['RELACAOFEM'].plot.hist(bins=100)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.distplot(df_brasil['RELACAOFEM'], bins=100, color='red', kde=False)
plt.title('Relacao eleitoras / eleitores', fontsize=18)
plt.xlabel('Eleitoras / eleitores', fontsize=14)
plt.ylabel('Frequencia', fontsize=14)
plt.axvline(1.0, color='black', linestyle='--')
```

```python
sns.swarmplot(data=df_brasil, x='NM_REGIAO', y='RELACAOFEM')
plt.axhline(1.0, color='black', linestyle='--')
```
